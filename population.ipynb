## **Loading CSV Data into a Pandas DataFrame**

The command `df = pd.read_csv('cities_r2.csv')` is used to load data from a CSV file into a Pandas DataFrame.

### Operation Explanation:
- **`pd.read_csv('cities_r2.csv')`**: This function reads the CSV file named `'cities_r2.csv'` and converts it into a DataFrame, which is a two-dimensional, labeled data structure similar to a table in a database or an Excel spreadsheet.
- **`df`**: This variable stores the resulting DataFrame, allowing you to perform various data analysis operations on it.

After executing this command, `df` will contain all the data from `'cities_r2.csv'`, and you can manipulate, analyze, or visualize this data using Pandas and other Python libraries.

````python
df = pd.read_csv('cities_r2.csv')
df
````

## **Removing a Column from a DataFrame**


This Python code snippet performs the following operations:

* **`df.drop(columns=['child_sex_ratio'])`**: This part of the code specifically targets the 'child_sex_ratio' column within the DataFrame `df` and removes it. The `drop` function is used for excluding specified data from the DataFrame.
* **`df_subset`**: The resulting DataFrame after the column removal is assigned to a new variable named `df_subset`. This new DataFrame will contain all columns from the original `df` except for 'child_sex_ratio'.


````python
df_subset = df.drop(columns=['child_sex_ratio'])
df_subset
````

## **Sorting a DataFrame by Column Values**

The `sort_values()` method in pandas allows you to sort a DataFrame based on the values of one or more columns. Here's how it works:

1. **DataFrame**: You start with a pandas DataFrame (`df_subset` in your case).
2. **Column to Sort By**: Specify the column by which you want to sort the DataFrame. In your example, you're sorting by the `'state_name'` column.
3. **Ascending or Descending Order**: By default, the sorting is done in ascending order (from smallest to largest). If you want to sort in descending order (largest to smallest), you can set the `ascending` parameter to `False`.
4. **Result**: The sorted DataFrame (`sort_df`) will have rows rearranged based on the values in the specified column.



````python
sort_df = df_subset.sort_values(by='state_name')
sort_df
````

## **Calculating Sex Ratio in a Pandas DataFrame**

1. **DataFrame**: You start with a pandas DataFrame (`df` in your case).
2. **New Column Creation**: You're creating a new column called `'sex_ratio'`.
3. **Calculation**: The values in the `'sex_ratio'` column are calculated by dividing the `'population_female'` by the `'population_male'` and then multiplying the result by 1000.
4. **Result**: The resulting `'sex_ratio'` column will contain sex ratio values (female per 1000 males) for each row in the DataFrame.


After executing this command, this DataFrame (`df`) will have an additional column called `'sex_ratio'` with the calculated sex ratio values.


````python
df['sex_ratio'] = (df['population_female'] / df['population_male']) * 1000
df
````

## **Bar Chart: Total Literates by State**

The command sequence generates a bar chart to visualize the total number of literates across different states. Here's a breakdown of the operation:

- `plt.figure(figsize=(14, 6))`: Sets the figure size to 14x6 inches for better readability.
- `plt.bar(df['state_name'], df['literates_total'], color='teal')`: Creates a bar chart with states on the x-axis and the total number of literates on the y-axis, with bars colored teal.
- `plt.xlabel('States')` and `plt.ylabel('Total Literates')`: Labels the x-axis as "States" and the y-axis as "Total Literates."
- `plt.title('Total Literates vs States')`: Sets the chart title to "Total Literates vs States."
- `plt.xticks(rotation=90)`: Rotates the x-axis labels by 90 degrees for better visibility.



````python
plt.figure(figsize=(14, 6))
plt.bar(df['state_name'], df['literates_total'], color='teal')
plt.xlabel('States')
plt.ylabel('Total Literates')
plt.title('Total Literates vs States')
plt.xticks(rotation=90)
plt.show()
````

## **Scatter Plot: Graduates vs. Effective Literacy Rate**


This set of commands performs a calculation and then visualizes the relationship between the total number of graduates and the effective literacy rate:

- `df['effective_literacy_rate'] = (df['literates_total'] / df['population_total']) * 100`: Calculates the effective literacy rate as a percentage by dividing the total number of literates by the total population for each state, then stores the result in a new column `effective_literacy_rate`.

- `plt.scatter(df['total_graduates'], df['effective_literacy_rate'], color='olivedrab', edgecolors='w', linewidth=0.2)`: Creates a scatter plot with the total number of graduates on the x-axis and the effective literacy rate on the y-axis. Data points are colored olive drab, with white edges and a thin outline.

- `plt.xlabel('Total Graduates')` and `plt.ylabel('Effective Literacy Rate')`: Labels the x-axis as "Total Graduates" and the y-axis as "Effective Literacy Rate."

- `plt.title('Total Graduates vs Effective Literacy Rate')`: Sets the chart title to "Total Graduates vs Effective Literacy Rate."


````python
df['effective_literacy_rate'] = (df['literates_total'] / df['population_total']) * 100
plt.figure(figsize=(10, 6))
plt.scatter(df['total_graduates'], df['effective_literacy_rate'], color='olivedrab',edgecolors='w', linewidth=0.2)
plt.xlabel('Total Graduates')
plt.ylabel('Effective Literacy Rate')
plt.title('Total Graduates vs Effective Literacy Rate')
plt.show()
````

##**KDE Plot: Distribution of Total Literates by State**

This command sequence produces a Kernel Density Estimate (KDE) plot to visualize the distribution of total literates for different states:

- `plt.figure(figsize=(18, 9))`: Initializes a large figure with dimensions of 18x9 inches.
- `sns.kdeplot(data=df, x='literates_total', hue='state_name', common_norm=False, fill='true')`: Generates a KDE plot where the x-axis represents the total literates. The `hue='state_name'` option creates separate density curves for each state, with `common_norm=False` allowing individual normalization, and `fill='true'` fills the area under the curves for emphasis.
- `plt.title('KDE Plot of Total Literates by State')`: Adds a title to the plot.
- `plt.xlabel('Total Literates')` and `plt.ylabel('Density')`: Labels the axes as "Total Literates" and "Density," respectively.
- `plt.show()`: Renders the KDE plot for viewing.

````python
plt.figure(figsize=(18, 9))
sns.kdeplot(data=df, x='literates_total', hue='state_name', common_norm=False, fill='true')
plt.title('KDE Plot of Total Literates by State')
plt.xlabel('Total Literates')
plt.ylabel('Density')
plt.show()
````

##**Bubble Plot: Population vs. Effective Literacy Rate**

This code genertes a bubble plot to illustrate the relationship between total population and effective literacy rate:

- `df['effective_literacy_rate'] = df['literates_total'] / df['population_total'] * 100`: Calculates the effective literacy rate as a percentage and adds it as a new column to the DataFrame.
- `plt.figure(figsize=(12, 8))`: Sets the figure size to 12x8 inches.
- `bubble_size = df['total_graduates'] / 1000`: Determines the size of each bubble, scaling it based on the number of total graduates, with a factor of 1000 to adjust bubble size for better visibility.
- `plt.scatter(df['population_total'], df['effective_literacy_rate'], s=bubble_size, alpha=0.8, c='black', edgecolors='w', linewidth=0.5)`: Creates a scatter plot with bubbles, where the x-axis represents total population, the y-axis represents effective literacy rate, and bubble size represents the number of graduates. The bubbles are black with white edges and a transparency level (`alpha`) of 0.8.

- `plt.grid(False)`: Disables the grid lines for a cleaner look.


````python
df['effective_literacy_rate'] = df['literates_total'] / df['population_total'] * 100

# Create the scatter plot
plt.figure(figsize=(12, 8))
bubble_size = df['total_graduates'] / 1000  # Adjust the scaling factor as needed
plt.scatter(df['population_total'], df['effective_literacy_rate'],s=bubble_size, alpha=0.8, c='black', edgecolors='w', linewidth=0.5)

plt.xlabel('Total Population')
plt.ylabel('Effective Literacy Rate (%)')
plt.title('Relationship between Total Population and Effective Literacy Rate')
plt.grid(False)

plt.show()
````

##**Multi-Plot Visualization: Analyzing Literacy and Population Data**



- **Figure Setup**:  
  `fig, axs = plt.subplots(2, 2, figsize=(20, 15))` initializes a 2x2 grid of subplots within a figure sized 20x15 inches.

- **Bar Plot** (`axs[0, 0]`):  
  Displays a bar plot of total literates by state using `sns.barplot()`. The x-axis labels (state names) are rotated for better readability. The plot shows the number of literates in each state, with the x-axis labeled "State Name" and the y-axis labeled "Total Literates."

- **Scatter Plot** (`axs[0, 1]`):  
  Visualizes the relationship between the total number of graduates and the effective literacy rate. The plot includes grid lines for clarity, with the x-axis labeled "Total Graduates" and the y-axis labeled "Effective Literacy Rate (%)."

- **KDE Plot** (`axs[1, 0]`):  
  Uses a Kernel Density Estimate (KDE) plot to display the distribution of total literates by state, with density curves filled for emphasis. The x-axis represents "Total Literates."

- **Bubble Plot** (`axs[1, 1]`):  
  Plots the relationship between total population and effective literacy rate, with bubble sizes representing the number of graduates. Bubbles are color-coded in black, with white edges. The plot has grid lines, with the x-axis labeled "Total Population" and the y-axis labeled "Effective Literacy Rate (%)."

- **Layout and Display**:  
  `plt.tight_layout()` optimizes spacing between the plots, and `plt.show()` displays the entire grid of visualizations.

````python
fig, axs = plt.subplots(2, 2, figsize=(20, 15))

# Bar Plot
sns.barplot(x='state_name', y='literates_total', data=df, color='teal', ci=None, ax=axs[0, 0])
axs[0, 0].set_xticklabels(axs[0, 0].get_xticklabels(), rotation=90, ha='right')
axs[0, 0].set_title('Total Literates by State')
axs[0, 0].set_xlabel('State Name')
axs[0, 0].set_ylabel('Total Literates')

# Scatter Plot
axs[0, 1].scatter(df['total_graduates'], df['effective_literacy_rate'], color='olivedrab',edgecolors='w', linewidth=0.2)
axs[0, 1].set_title('Total Graduates vs Effective Literacy Rate')
axs[0, 1].set_xlabel('Total Graduates')
axs[0, 1].set_ylabel('Effective Literacy Rate (%)')
axs[0, 1].grid(True)

# KDE Plot
sns.kdeplot(data=df, x='literates_total', hue='state_name', fill=True, ax=axs[1, 0])
axs[1, 0].set_title('KDE Plot of Literates Total by State')
axs[1, 0].set_xlabel('Total Literates')

# Bubble Plot
axs[1, 1].scatter(df['population_total'], df['effective_literacy_rate'],
                  s=df['total_graduates']/1000, alpha=0.8,c='black', edgecolors='w', linewidth=0.5)
axs[1, 1].set_title('Population vs Effective Literacy Rate with Graduates as Bubble Size')
axs[1, 1].set_xlabel('Total Population')
axs[1, 1].set_ylabel('Effective Literacy Rate (%)')
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
````

### Analyzing Gender Distribution in the Top 10 States

This script calculates and visualizes the gender distribution in the top 10 states by total population. It follows these steps:

1. **Compute Total Population**: The total population for each state is calculated by summing the male and female populations.
   ```python
   state_pop['total_population'] = state_pop['population_male'] + state_pop['population_female']
   ```

2. **Select Top 10 States**: The states are sorted by total population in descending order, and the top 10 states are selected.
   ```python
   top10_states = state_pop.sort_values(by='total_population', ascending=False).head(10)
   ```

3. **Calculate Gender Percentages**: For the top 10 states, the percentage of the male and female populations relative to the total population is computed.
   ```python
   top10_states['male_percentage'] = (top10_states['population_male'] / top10_states['total_population']) * 100
   top10_states['female_percentage'] = (top10_states['population_female'] / top10_states['total_population']) * 100
   ```

4. **Prepare Data for Heatmap**: The data is reorganized to display male and female percentages in a heatmap format.
   ```python
   heatmap_data = top10_states[['male_percentage', 'female_percentage']].T
   ```

5. **Create and Display Heatmap**: A heatmap is generated to visually compare the male and female population ratios across the top 10 states.
   ```python
   plt.figure(figsize=(17, 6))
   sns.heatmap(heatmap_data, annot=True, cmap='viridis', fmt=".1f")
   plt.title('Male vs Female Population Ratio in Top 10 States')
   plt.xlabel('State')
   plt.ylabel('Gender')
   plt.show()
   ```

````python
state_pop['total_population'] = state_pop['population_male'] + state_pop['population_female']

top10_states = state_pop.sort_values(by='total_population', ascending=False).head(10)

top10_states['male_percentage'] = (top10_states['population_male'] / top10_states['total_population']) * 100
top10_states['female_percentage'] = (top10_states['population_female'] / top10_states['total_population']) * 100

heatmap_data = top10_states[['male_percentage', 'female_percentage']].T

# Create the heatmap
plt.figure(figsize=(17, 6))
sns.heatmap(heatmap_data, annot=True, cmap='viridis', fmt=".1f")
plt.title('Male vs Female Population Ratio in Top 10 States')
plt.xlabel('State')
plt.ylabel('Gender')
plt.show()
````
